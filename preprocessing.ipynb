```python
def filter(datum) :
    x = datum["location_x"]
    y = datum["location_y"]
    if 100 < y and y < 170 and 830 < x and x < 870:
        datum["room"] = "stairs"
    elif 100 < y and y < 205 and 680 < x and x < 845:
        datum["room"] = "505"
    elif 125 < y and y < 183 and 845 < x and x < 927:
        datum["room"] = "503"
    elif 218 < y and y < 318 and 680 < x and x < 845:
        datum["room"] = "502"
    else :
        datum["room"] = "other"
    
    return datum
```
