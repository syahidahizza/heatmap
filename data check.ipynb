```python
df_all_data = pd.DataFrame.from_csv('../result/set9.csv').reset_index()
df_all_data = df_all_data[(df_all_data.location_y>200)&(df_all_data.location_x>580)]
df_all_data
```

```python
group = ['location_x', 'location_y', 'PCI']
df_data = merge_agg(df_all_data, group, 'RSRP', ['mean'])[group + ['mean']].drop_duplicates().reset_index(drop=True)
df_data = df_data.groupby(group[:-1]).filter(lambda x: x['PCI'].count() == 1).groupby(group).mean()
df = df_data.reset_index()
```

```python
df_data
```

```python
df = df.dropna()
lon_list = df["location_x"].astype('int32')
lat_list = df["location_y"].astype('int32')
rsrp_list = df["mean"].astype('int32')

rsrp_summary = summary_based_on_location(lat_list, lon_list, rsrp_list)
rsrp_summary = summary_dict(rsrp_summary, np.array)

normalize_rsrp_mean = matplotlib.colors.Normalize(vmin=min(rsrp_list), vmax=max(rsrp_list))
rsrp_mean = summary_dict(rsrp_summary, np.mean)
x_list, y_list, rsrp_mean_list = summary_dict_to_list(rsrp_mean)
colors_rsrp_mean = [cmap(normalize_rsrp_mean(value))[:3] for value in rsrp_mean_list]
colors_rsrp_mean = [[int(x*255) for x in value] for value in colors_rsrp_mean]

new_backtorgb = get_map_image()
new_backtorgb = visualize_cmap(new_backtorgb, x_list, y_list, colors_rsrp_mean, 
                               cmap, normalize_rsrp_mean, None)
```
